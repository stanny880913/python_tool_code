```python
from selenium import webdriver
import time
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
chrome_options = Options()
chrome_options.page_load_strategy = 'none'
driver = webdriver.Chrome(options=chrome_options)
# good url
driver.get('https://shop.gq.com.tw/product/5206')
```

### click to buy 

```python
while 1:
    try:
        if driver.current_url != 'https://shop.gq.com.tw/product/5206':
            print("yor are in loading page!!")
            driver.get('https://shop.gq.com.tw/product/5206')
        choose_color = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(
            (By.XPATH, '//*[@id="app"]/div[1]/main/div/section[1]/div/div[3]/div[3]/div[0]')))
        choose_color.click()

        choose_gray = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(
            (By.XPATH, '//*[@id="app"]/div[1]/main/div/section[1]/div/div[3]/div[3]/div[1]/div/ul/li[3]')))
        choose_gray.click()

        choose_size = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(
            (By.XPATH, '//*[@id="app"]/div[1]/main/div/section[1]/div/div[3]/div[3]/div[2]')))
        choose_size.click()

        choose_miedium_size = WebDriverWait(driver, 10).until(EC.element_to_be_clickable(
            (By.XPATH, '//*[@id="app"]/div[1]/main/div/section[1]/div/div[3]/div[3]/div[2]/div/ul/li[2]')))
        choose_miedium_size.click()
        
        buy_now = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CLASS_NAME, 'product-checkout__buy')))
        buy_now.click()

        break
    except:
        driver.refresh()
        print("not sale")
```
